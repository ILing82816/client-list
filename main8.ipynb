```python
import pandas as pd
main = pd.read_csv("company.csv", encoding= "utf-8")
dis = pd.read_csv("dissolution.csv", encoding= "utf-8")
```

```python
main['type'] = 'company'
main.set_index('type', append=True, inplace=True)
com = main.reset_index()
com
```

```python
dis['type'] = 'dissolution'
dis.set_index('type', append=True, inplace=True)
diss = dis.reset_index()
diss
```

```python
merged = com.append(diss)
merged.sort_values("公司名稱", inplace = True) 
merged
```

```python
merged.drop_duplicates(subset ="公司名稱", 
                     keep = False, inplace = True) 
merged
```

```python
result = pd.DataFrame()
result = (merged[merged["type"]=="company"].reset_index()
         .drop(["index","type", "level_0", "核准解散日期"], axis=1))
result.to_csv("client.csv", encoding= "utf_8", index=False)
result
```
