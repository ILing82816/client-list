```python
import requests
from bs4 import BeautifulSoup
href_list = []
title_list = []
response = requests.get("https://data.gov.tw/dataset/6047")
html = BeautifulSoup(response.text)
table = html.find("div", class_="field-name-field-dataset-resource")
month = table.find_all("a", class_="ff-icon-csv")
title = table.find_all("span", class_="ff-desc")
for m in month:
    href_list.append(m["href"])
for t in title:
    title_list.append(t.text)
```

```python
from urllib.request import urlretrieve
for h in range(len(title_list)):
    print(title_list[h], href_list[h])
    fname = "company/" + title_list[h] + ".csv"
    urlretrieve(href_list[h], fname)
```

```python
import glob
import pandas as pd
df = pd.DataFrame()
for fn in glob.glob("company/*.csv"):
    c = pd.read_csv(fn, encoding= "utf-8")
    loc = pd.Series([]) 
    for i in range(len(c)): 
        if "臺北市" in c["公司所在地"][i]: 
            loc[i]="臺北市"
        elif  "新北市" in c["公司所在地"][i]: 
            loc[i]="新北市"
        else: 
            loc[i]="其他"
    c.insert(4, "location", loc)
    df =pd.concat([df, c])
df.reset_index()
```

```python
result = pd.DataFrame()
result = df[df["location"].isin(["新北市", "臺北市"])]
result
```

```python
final = (result[(result["資本額"] >= 1000000)&(result["資本額"] <= 30000000)]
                .reset_index().drop(["index","序號"], axis=1))
final.to_csv("company.csv", encoding= "utf_8", index=False)
final
```
