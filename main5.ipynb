```python
import requests
from bs4 import BeautifulSoup
href_list = []
title_list = []
response = requests.get("https://data.gov.tw/dataset/6668")
html = BeautifulSoup(response.text)
table = html.find("div", class_="field-name-field-dataset-resource")
month = table.find_all("a", class_="ff-icon-csv")
title = table.find_all("span", class_="ff-desc")
for m in month:
    href_list.append(m["href"])
for t in title:
    title_list.append(t.text)
```

```python
from urllib.request import urlretrieve
for h in range(len(title_list)):
    print(title_list[h], href_list[h])
    fname = "list/" + title_list[h] + ".csv"
    urlretrieve(href_list[h], fname)
```

```python
import glob
import pandas as pd
df = pd.DataFrame()
for fn in glob.glob("list/*.csv"):
    c = pd.read_csv(fn, encoding= "utf-8")
    df =pd.concat([df, c])
df.reset_index()
```

```python
result = pd.DataFrame()
result = (df[(df["縣市名稱"] == "新北市")|(df["縣市名稱"] == "臺北市") ].reset_index()
          .drop(["index","序號"], axis=1))
result.to_csv("taipei.csv", encoding="utf-8", index=False)
result
```
